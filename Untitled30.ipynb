```python
from pyspark.sql import SparkSession
spark1 = SparkSession.builder.appName("Music_Emotion_Analysis").getOrCreate()
df = spark1.read.csv("/content/drive/MyDrive/top50.csv", header=True, inferSchema=True)
print("dataset avant la preparation ")
df.show(50)
```

```python
# Renommer les colonnes pour un format plus simple
df = df.withColumnRenamed("Track.Name", "track_name") \
       .withColumnRenamed("Artist.Name", "artist_name") \
       .withColumnRenamed("Beats.Per.Minute", "bpm") \
       .withColumnRenamed("Loudness..dB..", "loudness_db") \
       .withColumnRenamed("Valence.", "valence") \
       .withColumnRenamed("Length.", "length") \
       .withColumnRenamed("Acousticness..", "acousticness") \
       .withColumnRenamed("Speechiness.", "speechiness")

df.printSchema()
```

```python
colonnes_a_supprimer = ["bpm", "Danceability", "loudness_db", "Liveness", "acousticness", "loudness_db", "length", "speechiness"]
df = df.drop(*colonnes_a_supprimer)
df.printSchema()
```

```python
print("dataset après la prèparation : ")
df.show(5)
```

```python
from pyspark.sql.functions import when, col

df = df.withColumn(
    "emotion",
    when((col("valence") > 50) & (col("energy") > 50), "Happy")
    .when((col("valence") <= 50) & (col("energy") > 50), "Angry")
    .when((col("valence") > 50) & (col("energy") <= 50), "Relaxed")
    .otherwise("Sad")
)
from pyspark.sql.functions import when, col

df = df.withColumn(
    "season",
    when((col("energy") > 70) & (col("valence") > 50), "Summer")    # Haute énergie + Valence positive = Été
    .when((col("energy") > 50) & (col("valence") > 50), "Spring")   # Énergie modérée à élevée + Valence positive = Printemps
    .when((col("energy") <= 50) & (col("valence") <= 50), "Winter") # Énergie faible à modérée + Valence négative = Hiver
    .when((col("energy") <= 50) & (col("valence") > 50), "Autumn")  # Énergie faible à modérée + Valence positive = Automne
    .otherwise("Autumn")                                           # Par défaut, Automne (pour valence positive + énergie faible)
)

df.show(50)
```

```python
from pyspark.ml.recommendation import ALS  # Importation de l'algorithme de recommandation ALS
from pyspark.sql import SparkSession  # pour l'utilisation de Spark DataFrame
from pyspark.sql.functions import monotonically_increasing_id, rand, col, explode
from pyspark.ml.feature import StringIndexer  # pour transformer les noms en identifiants numériques

# Initialisation de la session Spark
spark = SparkSession.builder.appName("Recommendation_System1").getOrCreate()
# Chargement des données utilisateur-chanson
user_data = spark.read.csv("/chemin/de/sortie/processedtop50.csv", header=True)

# Ajout d'une colonne 'user_id' pour représenter des utilisateurs fictifs, avec des identifiants croissants
user_data = user_data.withColumn("user_id", monotonically_increasing_id())

# Transformation des noms de chansons en identifiants numériques avec StringIndexer
indexer = StringIndexer(inputCol="track_name", outputCol="song_id")  # 'track_name' devient 'song_id'
user_data = indexer.fit(user_data).transform(user_data)

# Ajout de colonne 'rating' avec des valeurs aléatoires entre 0 et 5 pour simuler les évaluations des users
user_data = user_data.withColumn("rating", (rand() * 5).cast("float"))

# Division des données en ensemble d'entraînement (80%) et ensemble de test (20%)
train, test = user_data.randomSplit([0.8, 0.2])

# Création du modèle de recommandation ALS avec les colonnes appropriées
als = ALS(
    userCol="user_id",
    itemCol="song_id",
    ratingCol="rating",
    nonnegative=True,
)

# Entraînement du modèle avec l'ensemble d'entraînement
model = als.fit(train)

# Génération de prédictions sur l'ensemble de test
predictions = model.transform(test)

# Génération de recommandations pour tous les utilisateurs (3 recommandations par utilisateur)
user_recommendations = model.recommendForAllUsers(3)

# Exploding de la colonne 'recommendations' pour convertir chaque recommandation en une ligne séparée
recommendations_exploded = user_recommendations.withColumn("recommendation", explode("recommendations"))

# Sélection des colonnes nécessaires : user_id, song_id et score (note de la recommandation)
recommendations_details = recommendations_exploded.select(
    "user_id",
    col("recommendation.song_id").alias("song_id"),
    col("recommendation.rating").alias("score")
)

# Jointure des recommandations avec les titres des chansons en utilisant 'song_id'
recommendations_with_titles = recommendations_details.join(
    user_data.select("song_id", "track_name").distinct(),
    on="song_id",
    how="inner"
)

# Affichage des recommandations avec les titres des chansons
print("Resultats du model de recommandations")
recommendations_with_titles.show(5, truncate=False)
```
